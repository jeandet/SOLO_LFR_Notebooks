# LFR BOOT Vs SUB_1V5 Voltage and Temperature

## Observed Problem

During thermal vacuum tests LFR was unable to boot over 55°C on Nominal link and 70°C on Redundant link. The boot process did fail while writing the first data into SRAM. The first 26 RMAP packets sent to LFR were acknowledged while the 27th packet wasn't. The difference is that the 27th packet does a SRAM write whereas the previous ones did only write LEON3 registers. 
As shown in the following schematic, the 1.8V power line supplying the SRAM core can be produced either from SUB_3V3 or SUB_1V5. A test have been performed on 06/08/2015 to ensure that the SUB_1V5 was sufficient to produce 1.8V. This tests was successful at room temperature with LFR EQM2 board (report PW-SYS-MEB-ELE-TN-001283-LES).

## Test goal
The aim of this test is to check LFR capacity to boot at high temperature with low SUB_1V5 voltages.

## Test Setup

![setup](./ressources/Setup.png)

### Instrument list
* Thermal chamber: Secasi **SLT 64/70**  + eruotherm **3508**
* SPW brick: StarDundee **USB SPW brick MK1**
* Thermal Sensors: [DLP-TEMP](http://www.dlpdesign.com/usb/temp.shtml) + 3x **DS18B20**
* LFR PSU: Custom power supply
* LAB PSU: GW INSTEK **GPD 3303S**
* I probe: Tektronix **AM503A** + **A6302**
* USB Scope: DreamSourceLab **DSLogic Oscillosocpe**

## Test Procedure

1. Connect all equipment as described in [setup](#Test-Setup)
2. Take pictures of the setup
3. Start LFR PSU recording
3. Check LFR PSU recording
4. Start temperature recording
5. Check temperature recording
6. Start LFR SGSE
7. Set temperature On thermal chamber controller to the temperature to test
8. Wait until LFR reach target temperature and is stable
9. Start USB scope and check that it waits for trigger
10. Launch launchtests_[**V**]V.sh replace **V** by the voltage to test
11. Connect LFR SGSE
12. Check if LFR sends packets
13. Save measured current with the USB scope as BOOT_CUR_[**T**]C_[**V**]V.txt replace **T** by temperature and **V** by the Voltage to test
14. Wait until the script exits
15. Repeat Step 10. or continue if no more volatge to test
16. Repeat Step 7. or continue if no more temperature step test
17. Turn OFF LFR PSU
18. Stop LFR PSU recording
19. Set temperature On thermal chamber controller to 30°C
20. Wait until LFR Temperature reach 30°C
21. Stop temperature recording
22. END OF TEST


* The temperature steps are **30°C**, **40°C**, **50°C**, **60°C**, **70°C** and **80°C**.
* The voltages to test are **2.2V**, **2.1V**, **2.05V**, **2.0V** and **1.9V**.

This should at least produce **110** steps and results.
