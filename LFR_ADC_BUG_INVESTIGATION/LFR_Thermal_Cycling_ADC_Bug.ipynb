# LFR ADC bug investigations

## Observed Problem

## Test goal

## Test Setup

![setup](./ressources/Setup.png)

## Instrument list
- Thermal chamber: Secasi ** SLT 64/70 ** + eruotherm ** 3508 **
- SPW brick: StarDundee ** USB SPW brick MK1 **
- Thermal Sensors: [DLP-TEMP](http://www.dlpdesign.com/usb/temp.shtml) + 3x ** DS18B20 ** 
- LFR PSU: Custom power supply
- LFR EGSE: Custom waveform generator

## Test Procedure

1. Modify LFR EQM board (chagne ADC clk config)
2. Take pictures of the modifications
3. Connect all equipment as described in setup
4. Take pictures of the setup
5. Start temperature recording
6. Check temperature recording
7. Start LFR EGSE
8. Power ON LFR /
9. **check power consumption**
10. Boot LFR
11. Configure Wveform generator to generate a **different** signal on each LFR input
12. Set Waveform Sapshot period to **22 secondes**
13. Start SPW packets recording
14. Enter **Normal** mode
15. Set **70°C*** as thermal chamber temperautre 
16. Wait until LFR board reach 70°C plus few minutes
17. Set **-50°C*** as thermal chamber temperautre 
18. Wait until LFR board reach -50°C plus few minutes
19. Set **70°C*** as thermal chamber temperautre 
20. Wait until LFR board reach 70°C plus few minutes
21. Set **30°C*** as thermal chamber temperautre 
22. Wait until LFR board reach 30°C
23. Turn OFF LFR
